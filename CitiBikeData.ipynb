```python
# List of file names to be combined 
files32023 = ['202308-citibike-tripdata.csv', '202309-citibike-tripdata.csv', '202310-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df32023_list = []

# Loop through each file in the list
for file in files32023:
    # Read the file into a dataframe
    df32023 = pd.read_csv(file)

    # Append the dataframe to the list
    df32023_list.append(df32023)

# Concatenate all dataframes in the list into one dataframe
df32023 = pd.concat(df32023_list)

# Show dataframe
df32023
```

```python
# Save the dataframe to a CSV file
df32023.to_csv('citibike.csv', index=False)
```
